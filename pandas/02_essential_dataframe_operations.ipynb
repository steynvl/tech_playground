## Transposing the direction of a DataFrame operation

```python
colleges = pd.read_csv("data/college.csv", index_col="INSTNM")
colleges.head(2)
```

```python
college_ugds = colleges.filter(like="UGDS_")
college_ugds.head(2)
```

```python
college_ugds.count(axis="columns").head()
```

```python
college_ugds.sum(axis="columns").head()
```

```python
college_ugds.cumsum(axis=1).head(2)
```

## Determine college campus diversity

```python
colleges = pd.read_csv("data/college.csv", index_col="INSTNM")
college_ugds = colleges.filter(like="UGDS_")
college_ugds.head(3)
```

```python
college_ugds.isnull().sum(axis=1).sort_values(ascending=False).head(4)
```

```python
college_ugds = college_ugds.dropna(how="all")
college_ugds.isnull().sum(axis=1).head(3)
```

```python
diversity_metric = college_ugds.ge(0.15).sum(axis=1)
diversity_metric.sort_values(ascending=False).head()
```
