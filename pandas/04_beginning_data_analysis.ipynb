## Selecting the smallest of the largest

```python
movies = pd.read_csv("data/movie.csv")
movies2 = movies[["movie_title", "imdb_score", "budget"]]
movies2.head(3)
```

```python
movies2.nlargest(100, "imdb_score").nsmallest(5, "budget")
```

## Selecting the largest of each group by sorting
#### Find the highest-rated film of each year

```python
movies[
    ["movie_title", "title_year", "imdb_score"]
].sort_values(["title_year", "imdb_score"], ascending=False).drop_duplicates(subset="title_year")
```

## Replicating nlargest with sort_values

```python
movies2.nlargest(100, "imdb_score").nsmallest(5, "budget")

movies2.sort_values("imdb_score", ascending=False, kind="mergsort").head(100).sort_values("budget", kind="mergsort").head(5)
```
